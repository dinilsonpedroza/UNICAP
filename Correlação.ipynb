# **CORRELAÇÃO**

A fórmula da correlação usa a covariância e o desvio-padrão. Vamos começar importando as bibliotecas de nosso exemplo.

```python
import numpy as np ### Trabalhar com números de forma organizada.
import pandas as pd ### Biblioteca básica para ciência de dados.
```

##**Covariância**

Agora vamos criar dados hipotéticos. Em que **X:** horas de estudo e **Y:** notas nas provas.

```python
# Horas de estudo
X = np.array([1, 2, 3, 4, 5]) ### Estamos criando um vetor, uma lista de valores.

# Notas
Y = np.array([3, 4, 5, 7, 8])

# Médias
media_X = np.mean(X) ### Aplicamos a função mean para encontrar a média da lista.
media_Y = np.mean(Y)

print("Média de X:", media_X) ### Imprimimos acrescentando o nome entre aspas.
print("Média de Y:", media_Y)
```

Vamos agora calcular a **covariância** "manualmente".

```python
cov = np.mean((X - media_X) * (Y - media_Y)) ### Confira com a fórmula mais acima.

print("Covariância:", cov) ### Não há intervalo de variação para a covariância.
```

Vamos ver em maior detalhes o resultado calculado.

```python
resultados_detalhados = pd.DataFrame({
    "X": X,
    "Y": Y,
    "X - média": X - media_X,
    "Y - média": Y - media_Y,
    "Produto": (X - media_X) * (Y - media_Y)
})

print(resultados_detalhados)
```

## **Desvio-padrão**

Precisamos dos desvios-padrão das horas estudadas (X) e das notas (Y). O primeiro vamos calcular "manualmente", o segundo usando a função std() da biblioteca Numpy.

```python
# Desvio-padrão manual
desvio_padrao_X = np.sqrt(np.mean((X - media_X)**2)) ### Aplicando a fórmula de desvio-padrão.

print("Desvio-padrão_X:", desvio_padrao_X) ### Apresentando o resultado.

# Desvio-padrão usando a função pronta do NumPy
print("Desvio-padrão_Y com NumPy:", np.std(Y))
```

Calculando a correção.

```python
desvio_padrao_Y = np.std(Y)
corr_X_Y = cov / (desvio_padrao_X * desvio_padrao_Y) ### Esse código foi proposto pelo Gemini intergardo ao Colab.

print("Correlação (X, Y):", corr_X_Y)
```

A correlação encontrada foi positiva e muito intensa. Muito próxima da linearidade absoluta, quando os pontos se situam em uma reta. Mas qual a diferença entre a covariância e a correlação? A covariância apenas mostra se as variáveis se movimentam juntas. E mais: o valor obtido pela fórmula varia de acordo com a unidade de medida das variáveis. A correlação padroniza a medida da relação entre as variáveis de duas formas: padroniza - ou normaliza - o resutado e mede sua intensidade: a correlação varia entre -1 e 1. Quanto mais próximo de 1, mais linear e positiva é a relação. Quanto mais próxima de -1, temos uma relação forte e negativa. A correlação está próxima de zero, então as duas variáveis não guardam relação entre si.
